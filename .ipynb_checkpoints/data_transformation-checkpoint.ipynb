```python
# Import Libraries
import numpy as np
import pandas as pd
```

```python
# Import CSV into dataframe
df = pd.read_csv('heart.csv',delimiter=',',header='infer')

# Display the first records in the dataframe
df.head()
```

```python
# Look at the summary stats for each column
df.describe()
```

```python
# Look at the shape of the dataframes
df.shape
```

```python
# Identify any columns with missing data
df.isnull().sum()
```

```python
# Identify if there are any duplicated rows
df.duplicated().any()
```

```python
# Print duplicated rows
df[df.duplicated()]
```

```python
# For each column print the unique values
for col in df.columns:
    print(col)
    print(df[col].unique())
    print('\n')
```

```python
# Change all the continous columns to categorical
df['age'] = pd.qcut(df['age'], q=4, labels=['very low', 'low', 'medium', 'high'])
df['trestbps'] = pd.qcut(df['trestbps'], q=4, labels=['very low', 'low', 'medium', 'high'])
df['chol'] = pd.qcut(df['chol'], q=4, labels=['very low', 'low', 'medium', 'high'])
df['thalach'] = pd.qcut(df['thalach'], q=4, labels=['very low', 'low', 'medium', 'high'])
df['oldpeak'] = pd.qcut(df['oldpeak'].rank(method='first'), q=4, labels=['very low', 'low', 'medium', 'high'])
```

```python
# Save the transformed dataset
df.to_csv('cleansed_heart.csv', index=False)
print('The csv was successfully saved as cleansed-healthcare-dataset-stroke-data.csv')
```
